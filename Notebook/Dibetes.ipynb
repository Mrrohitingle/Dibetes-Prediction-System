### Dependent and Independent Feature

```python
X=df.drop(columns='Outcome')
X
```

```python
y=df[['Outcome']]
```

```python
pd.DataFrame(y)
```

## Train Test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train_scaled,y_train)
```

```python
y_pred=classifier.predict(X_test_scaled)
y_pred
```

```python
classifier.predict_proba(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
model=LogisticRegression()
parameter = {
    'C': [ 0.1, 1, 10, 20],
    'penalty': ['l1', 'l2', 'elasticnet', 'none']
}
clf=GridSearchCV(estimator=model, param_grid=parameter,cv=5)
clf.fit(X_train_scaled,y_train)
```

```python
print(clf.best_params_)
```

```python
print(f"Best Estimator: {clf.best_estimator_}")
```

```python
best_classifier=clf.best_estimator_
best_classifier.fit(X_train_scaled,y_train)
y_pred=best_classifier.predict(X_test_scaled)
y_pred
```

```python
print(accuracy_score(y_test,y_pred))
```
